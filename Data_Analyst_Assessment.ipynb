```python
import numpy as np

df = df.assign(unit=lambda df: df["COMPANY FORMED"].str.extract("([a-z]+)+"),
         val=lambda df: df["COMPANY FORMED"].str.extract("([0-9,\.]+)").astype(float),
         months=lambda df: np.where(df["unit"].isin(["month","months"]), df["val"],
                                    np.where(df["unit"].isin(["year","years"])
                                             &df["val"].lt(55), df["val"]*12, np.nan)))
```

```python
import numpy as np
import pandas as pd

## Get All company Links
import requests
from bs4 import BeautifulSoup
```

```python
from sqlalchemy.sql.expression import null
CompanyID = []
for index, i in df.iterrows():
    # print(i['Company Name'])
    # if index>=100: #continue to start loop from second value
    #     continue
    req = requests.get('https://find-and-update.company-information.service.gov.uk/search?q='+ str(i['Company Name'])).text
    soup = BeautifulSoup(req, 'html.parser')
    link = soup.find(['h3', 'a'], class_='govuk-link',title='View company')
    cid = link.get('href') if link else ''
    CompanyID.append(cid)

    # if df['Company Name'].value[index] == link.get_text():
        # print(link.get('href'))
    # df['CompanyId'].values[index]=link.get("a")
```

```python
# CompanyID= CompanyID.replace(r'\s+|\\n', ' ', regex=True)
df['CompanyId'] =pd.Series(CompanyID)
# CompanyID
```
